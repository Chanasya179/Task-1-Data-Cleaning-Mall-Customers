```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
print("Shape:", df.shape)
df.head()
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("After removing duplicates:", df.shape)
```

```python
print("Before:", df['Gender'].unique())
df['Gender'] = df['Gender'].str.strip().str.title()
print("After:", df['Gender'].unique())
```

```python
print("Before:", df.columns.tolist())
df.columns = ['customer_id', 'gender', 'age', 'annual_income', 'spending_score']
print("After:", df.columns.tolist())
```

```python
df.dtypes
```

```python
print("===== FINAL CLEANED DATASET =====")
print("Shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())
print("\nData types:")
print(df.dtypes)
print("\nSample:")
df.head()
```

```python
df.to_csv("mall_customers_cleaned.csv", index=False)
```
